```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('Data Model - Pizza Sales.xlsx')
```

```python
df.columns
```

```python
df['order_dates'] = df['order_date'].dt.day_name()
df['order_dates']
g = df['order_dates'].value_counts()
h = list(g)
print(h)

df['order_dates'] = df['order_date'].dt.day_name()
i = df['order_dates'].unique()
j = list(i)
j
```

```python
from cycler import cycler

cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

barras = plt.bar(j, h)

plt.bar_label(barras, labels=h, color='green')
plt.xticks(rotation=90)
plt.xlabel('Datas', fontsize=12, color='green')
plt.ylabel('Frequency', fontsize=12, color='green')
plt.title('Datas', color='green')

plt.tick_params(axis='x', length=0)



for i, barra in enumerate (barras):
    if i != h.index(max(h)):
        barras[i].set_color(cores[1])
        barras[i].set_alpha(0.3)

plt.show()
```
